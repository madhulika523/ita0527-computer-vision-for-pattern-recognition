```python
import cv2
import matplotlib.pyplot as plt

# Path to your image file
image_path = '/content/image 1.jpg'  # Updated to use the uploaded image

# Read the image
# cv2.imread returns None if the image cannot be read (e.g., wrong path, corrupt file)
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not read the image from {image_path}. Please check the path.")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Display the original and grayscale images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for matplotlib
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
    plt.imshow(gray_image, cmap='gray') # Display grayscale image with 'gray' colormap
    plt.title('Grayscale Image')
    plt.axis('off')

    plt.show()
```
