The code below will:
1. Read the image using OpenCV.
2. Convert the image from BGR to RGB (as OpenCV reads images in BGR format, but Matplotlib expects RGB).
3. Apply a Gaussian blur using `cv2.GaussianBlur`.
4. Display both the original and the blurred images using Matplotlib.

```python
import cv2
import matplotlib.pyplot as plt

# --- Configuration --- #
# Replace 'your_image_name.jpg' with the path to your image file.
# For example, if you uploaded 'my_picture.png', change it to 'my_picture.png'
image_path = '/content/image 2.jpg'

# Gaussian Blur kernel size (odd numbers only, e.g., 5, 15, 25)
# Larger values result in more blur.
kernel_size = (15, 15)

# --- Image Processing --- #

# 1. Read the image
# cv2.imread returns None if the image cannot be read (e.g., file not found)
original_image_bgr = cv2.imread(image_path)

if original_image_bgr is None:
    print(f"Error: Could not read the image from {image_path}. Please check the path and ensure the image exists.")
else:
    # 2. Convert BGR image to RGB for displaying with matplotlib
    original_image_rgb = cv2.cvtColor(original_image_bgr, cv2.COLOR_BGR2RGB)

    # 3. Apply Gaussian Blur
    # The third argument is the standard deviation in X and Y direction respectively.
    # If 0, it's calculated from kernel size.
    blurred_image_rgb = cv2.GaussianBlur(original_image_rgb, kernel_size, 0)

    # --- Display Results --- #

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off') # Hide axes ticks

    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    plt.imshow(blurred_image_rgb)
    plt.title(f'Blurred Image (Kernel: {kernel_size})')
    plt.axis('off') # Hide axes ticks

    plt.tight_layout()
    plt.show()
```
