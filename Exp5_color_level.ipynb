```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def analyze_image_histogram(image_path):
    """
    Analyzes and displays the color histogram of an input image.

    Args:
        image_path (str): The path to the input image file.
    """
    # Load the image
    img = cv2.imread(image_path)

    if img is None:
        print(f"Error: Could not load image from {image_path}")
        return

    # Split the image into its B, G, R channels
    b, g, r = cv2.split(img)

    # Calculate histogram for each channel
    hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
    hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

    # Plot the histograms
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for correct display
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.plot(hist_b, color='blue', label='Blue Channel')
    plt.plot(hist_g, color='green', label='Green Channel')
    plt.plot(hist_r, color='red', label='Red Channel')
    plt.title('Color Channel Histograms')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Example Usage:
# For demonstration, let's create a dummy image first.
# In a real scenario, you would replace this with the path to your actual image file.

# Create a dummy image (e.g., 100x100 white image with a red square)
width, height = 100, 100
dummy_image = np.zeros((height, width, 3), dtype=np.uint8) # Black image
dummy_image[:, :] = [255, 255, 255] # Make it white (BGR)

# Draw a blue square (0,0,255 BGR) in the center
dummy_image[25:75, 25:75] = [255, 0, 0]

dummy_image_path = 'dummy_image.png'
cv2.imwrite(dummy_image_path, dummy_image)

print(f"Dummy image created at: {dummy_image_path}")

# Now, call the function with the path to the dummy image
analyze_image_histogram(dummy_image_path)

# To analyze your own image, replace 'dummy_image.png' with the path to your image file, e.g.:
# analyze_image_histogram('path/to/your/image.jpg')
```
