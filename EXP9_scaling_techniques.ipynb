```python
import cv2, matplotlib.pyplot as plt

img = cv2.imread('/content/image6.jpg')
if img is None: print('Error: Could not load image.'); exit()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

small_size = (int(img.shape[1] * 0.5), int(img.shape[0] * 0.5))
large_size = (int(img.shape[1] * 1.5), int(img.shape[0] * 1.5))

img_small = cv2.resize(img_rgb, small_size, interpolation=cv2.INTER_AREA)
img_large = cv2.resize(img_rgb, large_size, interpolation=cv2.INTER_LINEAR)

plt.figure(figsize=(15, 5))
plt.subplot(131); plt.imshow(img_rgb); plt.title('Original'); plt.axis('off')
plt.subplot(132); plt.imshow(img_small); plt.title('Smaller (50%)'); plt.axis('off')
plt.subplot(133); plt.imshow(img_large); plt.title('Larger (150%)'); plt.axis('off')
plt.show()
```
