```python
import cv2
import matplotlib.pyplot as plt

# --- Step 1: Read an image ---
# Replace 'path/to/your/image.jpg' with the actual path to your image file
# For example, if your image is in the same directory as your notebook, you can just use its filename like 'my_image.jpg'
image_path = '/content/image3.jpg'
original_image = cv2.imread(image_path)

# Check if the image was loaded successfully
if original_image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert the image from BGR to RGB for displaying with matplotlib
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # --- Step 2: Convert the Image to show outline using Canny function ---
    # Convert the image to grayscale (Canny usually works on grayscale images)
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

    # Apply Canny edge detector
    # The parameters are (image, threshold1, threshold2)
    # threshold1 and threshold2 are used for hysteresis thresholding
    edges = cv2.Canny(gray_image, 100, 200)

    # Display the original and edge-detected images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(original_image_rgb)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Canny Edge Detection')
    plt.imshow(edges, cmap='gray')
    plt.axis('off')

    plt.show()
```
