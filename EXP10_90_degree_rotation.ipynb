```python
import cv2, matplotlib.pyplot as plt

# Load the image (assuming image 2.jpg is now available)
img = cv2.imread('/content/image 2.jpg')

if img is None: print('Error: Could not load image.'); exit()

# Convert BGR to RGB for correct display with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Perform 90-degree clockwise rotation
# Note: cv2.ROTATE_90_CLOCKWISE performs a 2D rotation in the image plane.
# A 'rotation along the y-axis' usually implies a 3D perspective transform,
# but given the context and line limit, a 2D rotation is the common interpretation.
rotated_img = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)

# Display original and rotated images
plt.figure(figsize=(10, 5))
plt.subplot(121); plt.imshow(img_rgb); plt.title('Original Image'); plt.axis('off')
plt.subplot(122); plt.imshow(rotated_img); plt.title('Rotated 90° Clockwise'); plt.axis('off')
plt.show()
```
